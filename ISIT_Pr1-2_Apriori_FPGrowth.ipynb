```python
from apriori_python import aprioriFromFile
import pandas as pd

data = pd.read_csv('BreadBasket_DMS.csv', delimiter=',')
data = data.drop("Date", axis=1)
data = data.drop("Time", axis=1)

#data = data[data.Item != 'NONE']

#array = [[] * 1] * (data.Transaction.max() - data.Transaction.min() + 1)
array = []
for i in range(data.Transaction.max() - data.Transaction.min() + 1):
    array.append([]*1)

for i in range(data.shape[0]):
  if (data.Item[i] == 'NONE'):
    continue
  if data.Item[i] not in array[data.Transaction[i]-1]:
    array[data.Transaction[i]-1].append(data.Item[i])
```
