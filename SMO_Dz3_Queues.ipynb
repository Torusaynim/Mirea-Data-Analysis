```python
import math
import numpy as np
import matplotlib.pyplot as plt

λ = float(input("Введите интенсивность входящего потока λ: "))
μ = float(input("Введите интенсивность потока обслуживаний μ: "))
m = float(input("Введите ограничение на длину очереди m: "))
ρ = λ/μ
P_otk, L_och, T_och = 0, 0, 0
print("\nИнтенсивность нагрузки канала: ",ρ)

if (ρ != 1):
  P_otk = ((ρ**(m+1))*(1-ρ))/(1-(ρ**(m+2)))
  L_och = ((ρ**2)*(1-((ρ**(m))*(m+1-m*ρ))))/((1-ρ)*(1-(ρ**(m+2))))
else:
  P_otk = 1/(m+2)
  L_och = (m*(m+1))/(2*(m+2))

T_och = L_och / λ
print("Вероятность отказа: ", P_otk, "\nСреднее число заявок в очереди", L_och, "\nСреднее время ожидания заявки в очереди", T_och)
```
