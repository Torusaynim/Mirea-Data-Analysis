```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import seaborn as sns

data = pd.read_csv('insurance.csv', sep=",")
#print(data.head(5))
print(data.describe())

bminp = np.array(data["bmi"])
chargesnp = np.array(data["charges"])

bmiMean = np.mean(bminp)
bmiModa = sts.mode(bminp)
bmiMed = np.median(bminp)
bmiRaz = bminp.max() - bminp.min()
bmiStd = bminp.std()
bmiQ1 = np.percentile(bminp, 25, interpolation='midpoint')
bmiQ3 = np.percentile(bminp, 75, interpolation='midpoint')
bmiIqr1 = bmiQ3 - bmiQ1
bmiIqr2 = sts.iqr(bminp, interpolation='midpoint')

chargesMean = np.mean(chargesnp)
chargesModa = sts.mode(chargesnp)
chargesMed = np.median(chargesnp)
chargesRaz = chargesnp.max() - chargesnp.min()
chargesStd = chargesnp.std()
chargesQ1 = np.percentile(chargesnp, 25, interpolation='midpoint')
chargesQ3 = np.percentile(chargesnp, 75, interpolation='midpoint')
chargesIqr1 = chargesQ3 - chargesQ1
chargesIqr2 = sts.iqr(chargesnp, interpolation='midpoint')

fig, ax = plt.subplots(1, 4, figsize = (20, 5))
ax[0].hist(data["age"], edgecolor = 'black', color='green', bins=25)
ax[1].hist(data["bmi"], edgecolor = 'black', color='green', bins=25)
ax[2].hist(data["children"], edgecolor = 'black', color='green', bins=25)
ax[3].hist(data["charges"], edgecolor = 'black', color='green', bins=25)
ax[1].axvline(bmiMean, color='b', linestyle='dashed', linewidth=2, label = 'bmiMean')
ax[1].axvline(bmiModa[0], color='y', linestyle='dashed', linewidth=2, label = 'bmiModa')
ax[1].axvline(bmiMed, color='r', linestyle='dashed', linewidth=2, label = 'bmiMed')
#ax[1].axvline(bmiRaz, color='c', linestyle='dashed', linewidth=2, label = 'bmiRaz')
#ax[1].axvline(bmiStd, color='m', linestyle='dashed', linewidth=2, label = 'bmiStd')
#ax[1].axvline(bmiIqr1, color='k', linestyle='dashed', linewidth=2, label = 'bmiIqr1')
#ax[1].axvline(bmiIqr2, linestyle='dashed', linewidth=2, label = 'bmiIqr2')
ax[1].legend()
ax[3].axvline(chargesMean, color='b', linestyle='dashed', linewidth=2, label = 'chargesMean')
ax[3].axvline(chargesModa[0], color='y', linestyle='dashed', linewidth=2, label = 'chargesModa')
ax[3].axvline(chargesMed, color='r', linestyle='dashed', linewidth=2, label = 'chargesMed')
#ax[3].axvline(chargesRaz, color='c', linestyle='dashed', linewidth=2, label = 'chargesRaz')
#ax[3].axvline(chargesStd, color='m', linestyle='dashed', linewidth=2, label = 'chargesStd')
#ax[3].axvline(chargesIqr1, color='k', linestyle='dashed', linewidth=2, label = 'chargesIqr1')
#ax[3].axvline(chargesIqr2, linestyle='dashed', linewidth=2, label = 'chargesIqr2')
ax[3].legend()
plt.show()

print('Индекс массы тела:\nСреднее = ', bmiMean, '\nМода = ', bmiModa, '\nМедиана = ', bmiMed, '\nРазмах = ', bmiRaz, '\nСтандартное отклонение: ', bmiStd, '\nМежинтервальный размах через numpy = ', bmiIqr1, '\nМежинтервальный размах через scipy = ', bmiIqr2, '\n')
print('Расходы:\nСреднее = ', chargesMean, '\nМода = ', chargesModa, '\nМедиана = ', chargesMed, '\nРазмах = ', chargesRaz, '\nСтандартное отклонение: ', chargesStd, '\nМежинтервальный размах через numpy = ', chargesIqr1, '\nМежинтервальный размах через scipy = ', chargesIqr2, '\n')

plt.figure(figsize = (8, 8))
plt.suptitle('age', fontsize=20)
plt.boxplot(data["age"], vert = False)
plt.grid()
plt.show()

plt.figure(figsize = (8, 8))
plt.suptitle('bmi', fontsize=20)
plt.boxplot(data["bmi"], vert = False)
#plt.xticks(np.arange(0, len(data["bmi"]), 300))
plt.grid()
plt.show()

plt.figure(figsize = (8, 8))
plt.suptitle('children', fontsize=20)
plt.boxplot(data["children"], vert = False)
plt.grid()
plt.show()

plt.figure(figsize = (8, 8))
plt.suptitle('charges', fontsize=20)
plt.boxplot(data["charges"], vert = False)
#plt.xticks(np.arange(0, len(data["bmi"]), 300))
plt.grid()
plt.show()
```

```python
sts.probplot(data['bmi'], dist='norm', plot=plt)
plt.show()
print('\n')
sts.probplot(data['charges'], dist='norm', plot=plt)
plt.show()

loc, scale = sts.norm.fit(data['bmi'])
n = sts.norm(loc=loc, scale=scale)
bmiKStest = sts.kstest(data['bmi'], n.cdf)
print(bmiKStest)

loc, scale = sts.norm.fit(data['charges'])
n = sts.norm(loc=loc, scale=scale)
chargesKStest = sts.kstest(data['charges'], n.cdf)
print(chargesKStest)
```
