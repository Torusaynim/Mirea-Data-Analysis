```python
import math
import numpy as np
import matplotlib.pyplot as plt

def p0_func(ρ, n):
  if (n>=1):
    return (ρ**n)/(math.factorial(n)) + p0_func(ρ, n-1)
  else:
    return 1

λ = float(input("Input λ: "))
μ = float(input("Input μ: "))
ρ = λ/μ

n=1
Q=0
ResultArray = []
NArray = []

while Q<0.9:
  NArray.append(n)
  Q = 1 - ((ρ**n)/math.factorial(n))*(p0_func(ρ, n))**(-1)
  print("Q({}) = {}".format(n, Q))
  n = n+1
  ResultArray.append(Q)

print("Оптимальное число каналов для заданных λ и μ:", (n-1))
plt.plot(NArray, ResultArray, '-o', color='yellowgreen')
plt.yticks(np.arange(0, 1.01, 0.1))
plt.show()
```
